```python
# Import packages for project 
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#%matplotlib inline
#matplotlib.rcParams['figure.figsize'] = (12, 8)

pd.options.mode.chained_assignment = None
```

```python
# read in data
df = pd.read_csv("movies.csv")
```

```python
# Preview the data
df.head()
```

```python
# Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# delete rows with NaN values in any column

new_df = df.dropna(axis=0)

new_df.head()
```

```python
# Data types for columns
df.dtypes
```

```python
# change budget and gross columns into ints, gets rid of trailing .0
# ***  Int64 not int64
df['budget'] = df['budget'].astype('Int64')

df['gross'] = df['gross'].astype('Int64')

# ^chang^ add in votes column
```

```python
# create new column - updated_year
# ^change^ this, different from tutorial
df['updated_year'] = df['released'].astype(str).str[:4]

df.head()
```

```python
# Sort by gross revenue
most_gross = df.sort_values(by=['gross'], inplace=False, ascending = False)

most_gross.head()
```

```python
# if want to unlock limit on data shown in tables use
# pd.set_option('display.max_rows', None)
print('Notes for future use')
```

```python
# Drop any duplicates

#this shows unique values in set
df['company'].drop_duplicates().sort_values(ascending=False)

#this would drop all duplicates within table
#df = df.drop_duplicates()
```

```python
# test budget in comparison with company and gross profit
#create scatterplot, using new_data with no null values
plt.scatter(x=new_df['budget'], y=new_df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
#using seaborn, plot vs budget

sns.regplot(x='budget', y='gross', data=new_df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#Data correlation, pearson(using) vs kendall vs spearman
df.corr(method='pearson')
```

```python
# high correlation between budget and goss
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Looking at Company

df_numerized = most_gross

for col in df_numerized.columns:
    if(df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
#destacking
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
